# might be nice to look at stays as weekend, weeks, months... vs just days

```python
#minimum stay

min_stay = {
    # Weekend
    range(1, 4): "weekend",
    # Week
    range(4, 8): "week",
    # Month
    range(8, 31): "month",
    # 2 Months
    range(31, 61): "2_months",
     # 3 Months
    range(61, 91): "3_months",
     # 6 Months
    range(91, 181): "6_months",
     # Year
    range(181, 366): "year",
    # long term
    range(366,10001): "long_term"
}

# Use the map to create a new column for season
listing["min_stay_group"] = listing['minimum_nights'].apply(lambda day: next((stay for day_range, stay in min_stay.items() if day in day_range), None))
listing["min_stay_group"].value_counts(normalize=True)
```

```python
# We can also use the same map to make groups for maximum stay
listing["max_stay_group"] = listing['maximum_nights'].apply(lambda day: next((stay for day_range, stay in min_stay.items() if day in day_range), None))
listing["max_stay_group"].value_counts(normalize=True)
```

```python
#a few entries that can't be satisfied by a function

listing['property_type'].replace('Lighthouse', 'Entire home', inplace=True)
listing['property_type'].replace('Castle', 'Entire home', inplace=True)
listing['property_type'].replace('Tiny home', 'Entire home', inplace=True)
listing['property_type'].replace('Camper/RV', 'Entire home', inplace=True)

#i'm going to say a hotel space is a entire home, as the renter has the space to themselves. 

listing['property_type'].replace('Room in aparthotel', 'Hotel room', inplace=True)
listing['property_type'].replace('Room in boutique hotel', 'Hotel room', inplace=True)
listing['property_type'].replace('Room in hotel', 'Hotel home', inplace=True)

```

```python
listing['property_type'].value_counts()
```

```python
def accommodation_type(description):
    if pd.isnull(description):
        return 'missing'
    if 'Entire' in description:
        return 'entire home'
    elif 'Shared' in description:
        return 'shared space'
    elif 'Hotel' in description:
        return 'hotel Room'
    else:
        return 'private room'

listing['shared_status'] = listing['property_type'].apply(accommodation_type)
```
