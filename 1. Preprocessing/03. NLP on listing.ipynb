# NP analysis on written columns
I want to run sentiment analysis on the name, host_about, description and neighborhood_overview columns to be able to use tone as a factor.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# listings = pd.read_csv('../data/mergedAPI.csv')
listings = pd.read_csv('../data/mergedAPI2.csv')
```

```python
listings.head()
```

```python
#doing sentiment analysis on words, there have to be words. Neutral is not weighted either direction.  
listings['name'].fillna('blank', inplace=True)
listings['host_about'].fillna('blank', inplace=True)
listings['neighborhood_overview'].fillna('blank', inplace=True)
listings['description'].fillna('blank', inplace=True)
```

```python
sia = SentimentIntensityAnalyzer()

# Iterate over the rows of the DataFrame
for index, row in listings.iterrows():
    # Apply VADER sentiment analysis to the comment
    sentiments = sia.polarity_scores(row['name'])
    # Update the sentiment scores for this row
    listings.at[index, 'name_sent_compound'] = sentiments['compound']
    listings.at[index, 'name_sent_pos'] = sentiments['pos']
    listings.at[index, 'name_sent_neg'] = sentiments['neg']
    listings.at[index, 'name_sent_neu'] = sentiments['neu']
```
