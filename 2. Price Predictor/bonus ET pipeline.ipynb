I will make a pipeline, then gridsearch to see if I can get an even better model by adjusting the parameters.  

```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_squared_error

# import sklearn.preprocessing as preprocessing
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

import pickle
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Import

```python
df = pd.read_csv('../data/1500_sentiment_api.csv')
df.head(5)
```

```python
df.columns
```

### Creating a model

```python
X = df[[
    'shared_status', 
    'accommodates',
    'bathrooms', 
    'bedrooms',
    'beds_adjusted',
    'neighborhood', 
    'has_neighborhood_overview', 
    'has_reviews',
    'instant_bookable',
    'host_lives_in_neighborhood',
     'host_is_superhost', 
    'host_has_profile_pic',
    'host_identity_verified', 
    'has_host_about', 
    'years_hosting',
    'license_listed', 
    'minimum_nights',
    'maximum_nights',
     'number_of_reviews',
     'bike_scores',
    'walk_scores',
    'transit_scores',
    'host_sent_compound', 
    'description_sent_compound', 
    'neighborhood_sent_compound']]
    
    


y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2022)
```

```python
numeric_features = list(X.select_dtypes(include = 'number').columns)
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_features = list(X.select_dtypes(include = 'object').columns)
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Append classifier to preprocessing pipeline.
# Now we have a full prediction pipeline.
pipe = Pipeline(steps=[('preprocessor', preprocessor),
                      ('model', ExtraTreesRegressor())])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
rmse = (mean_squared_error(y_test, y_pred))**0.5
print("RMSE:", rmse)
```

```python
pipe.get_params()
```

```python
et_params = {
    'model__n_estimators':[100, 150, 200],
    'model__max_depth': [None, 1, 2, 3, 4, 5, 6,7],
    'model__min_samples_leaf' : [1,2]
}
```

```python
gs= GridSearchCV(pipe, param_grid=et_params, cv=5)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
y_pred = gs.predict(X_test)
rmse = (mean_squared_error(y_test, y_pred))**0.5
print("RMSE:", rmse)
```

```python
et_params2 = {
    'model__n_estimators':[150],
    'model__max_depth': [None],
    'model__min_samples_leaf' : [2],
    'model__max_features': ['sqrt', None],
    'model__min_weight_fraction_leaf': [x / 10 for x in range(0, 2)]
}
```

```python
gs= GridSearchCV(pipe, param_grid=et_params2, cv=5)
```

```python
gs.fit(X_train, y_train)
```

```python
y_pred = gs.predict(X_test)
rmse = (mean_squared_error(y_test, y_pred))**0.5
print("RMSE:", rmse)
```

```python
et_params3 = {
    'model__n_estimators':[150],
    'model__max_depth': [None],
    'model__min_samples_leaf' : [2],
    'model__max_features': ['sqrt', 'log2'],
    'model__min_weight_fraction_leaf': [0]
}
```

```python
gs= GridSearchCV(pipe, param_grid=et_params3, cv=5)
```

```python
gs.fit(X_train, y_train)
```

```python
y_pred = gs.predict(X_test)
rmse = (mean_squared_error(y_test, y_pred))**0.5
print("RMSE:", rmse)
```
